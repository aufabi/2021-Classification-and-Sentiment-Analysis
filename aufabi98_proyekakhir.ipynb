Setelah fungsi dipanggil dataframe yang direturn saya append ke list dataframe untuk kemudian digabungkan menggunakan method concat dan disimpan ke variabel df_mentah.

```python
# Membuat list df kosong yang nantinya akan diisi dengan hasil setiap df yang dicari
# Membuat fungsi untuk mengambil data pada tanggal tertentu
list_df = []
def cari_tweet(date):
    word = "ppkm"
    new_search = word + " -filter:retweets"
    tweets = tweepy.Cursor(api.search_tweets, q = new_search, lang = "id", until = date).items(50)
    items = []
    date_list = []
    for tweet in tweets:
        items.append (' '.join(re.sub("(@[A-Za-z0-9]+)|([^0-9A-Za-z \t])|(\w+:\/\/\S+)|(\d+)", " ", tweet.text).split()))
        date_list.append(word_tokenize(str(tweet.created_at))[0])
    df = pd.DataFrame(data=date_list, columns=['date'])
    df['tweet'] = items
    list_df.append(df)

# Memanggil fungsi
cari_tweet("2021-12-04")
cari_tweet("2021-12-05")
cari_tweet("2021-12-06")
cari_tweet("2021-12-07")

# Menggabungkan df di list df dengan concat
df_mentah = pd.concat([list_df[0], list_df[1]], ignore_index = True, axis = 0)
df_mentah = pd.concat([df_mentah, list_df[2]], ignore_index = True, axis = 0)
df_mentah = pd.concat([df_mentah, list_df[3]], ignore_index = True, axis = 0)

# Menampilkan data
df_mentah
```

## Preprocessing

Pada tahap ini, saya akan membersihkan tweet tadi dari angka, stop word, tanda baca, kata slang, dan lain sebagainya. Tweet yang bersih kemudian disimpan ke dalam kolom baru bernama clean tweet. Pertama - tama saya akan membuat beberapa fungsi untuk melakukan pembersihan data.

```python
# Melakukan pembersihan case folding : pengecilan huruf, tanda baca, spacing, whitespace, dll
def case_folding(data):
    data = data.lower()
    data = ' '.join(re.sub("(@[A-Za-z0-9]+)|([^0-9A-Za-z \t])|(\w+:\/\/\S+)|(\d+)"," ",data).split())
    data = data.translate(str.maketrans("","",string.punctuation))
    data = re.sub(r"\n","",data)
    data = re.sub(r"\t","",data)
    return data

# Membersihkan stopword pada bahasa Indonesia
def stopword_cleaner(data):
    data = word_tokenize(data)
    sw_indonesia = stopwords.words("indonesian")
    data  = [word for word in data if word not in sw_indonesia]
    data = ' '.join(data)
    return data

# Menghilangkan kata - kata singkatan dan slang
def slang_cleaner(data):
    slang_list = [
        ('yg', 'yang'), ('jd', 'jadi'), ('aja', 'saja'), ('ya', 'iya'), ('gak', 'tidak'), ('ga', 'tidak'), ('kalo', 'kalau'),
        ('klo', 'kalau'), ('jg', 'juga'), ('udah', 'sudah'), ('dgn', 'dengan'), ('kau', 'kamu'), ('tdk', 'tidak'), 
        ('lu', 'kamu'), ('biar', 'supaya'), ('nuntut', 'menuntut'), ('sj', 'saja'), ('gede', 'besar'), ('sy', 'saya'),
        ('krn', 'karena'), ('emang', 'memang'), ('sdh', 'sudah'), ('jt', 'juta'), ('ngga', 'tidak'), ('gk', 'tidak'),
        ('bikin', 'membuat'), ('nggak', 'tidak'), ('pake', 'memakai'), ('ngomong', 'berbicara'), ('sampe', 'sampai'),
        ('ngak', 'tidak'), ('org', 'orang'), ('utk', 'untuk'), ('n', 'dan'), ('blm', 'belum'), ('jgn', 'jangan'), 
        ('gini', 'begini'), ('jateng', 'jawa tengah'), ('jatim', 'jawa timur'), ('knp', 'kenapa'), ('tp', 'tapi'), 
        ('kayak', 'seperti'), ('gw', 'saya'), ('dg', 'dengan'), ('elu', 'kamu'), ('kemaren', 'kemarin'), ('k', 'ke'),
        ('doang', 'hanya'), ('pakai', 'memakai'), ('mikir', 'berfikir'), ('tetep', 'tetap'), ('bener', 'benar'), 
        ('ama', 'bersama'), ('ndak', 'tidak'), ('mesti', 'pasti'), ('jabar', 'jawa barat'), ('skrg', 'sekarang'), 
        ('udh', 'udah'), ('pengen', 'ingin'), ('naikin', 'menaikkan'), ('hny', 'hanya'), ('sklgs', 'sekaligus'), 
        ('lbh', 'lebih'), ('sgt', 'sangat'), ('bgt', 'sangat'), ('gbt', 'gabut'), ('liat', 'lihat'), ('mara', 'marah'),
        ('dadi', 'menjadi'), ('saiki', 'sekarang'), ('karna', 'karena'), ('udah', 'sudah'), ('smg', 'semoga'), 
        ('alm', 'almarhum'), ('banget', 'sangat'), ('melulu', 'selalu'), ('jdi', 'jadi'), ('nesu', 'marah'),
        ('walkot', 'wali kota'), ('njenengan', 'anda'), ('lg', 'sedang'), ('why', 'kenapa'), ('bs', 'bisa'), 
        ('kelakuane', 'perilakunya'), ('inget', 'ingat'), ('dikit', 'sedikit'), ('sm', 'sama'), ('twit', 'tweet'),
        ('mbela', 'membela'), ('ato', 'atau'), ('mngobati', 'mengobati'), ('moga', 'semoga'), ('kdatangan', 'kedatangan'),
        ('pnghibur', 'penghibur'), ('marahin', 'memarahi'), ('ttg', 'tentang'), ('mantau', 'memantau'), ('tulung', 'tolong'),
        ('ojo', 'jangan'), ('manteb', 'mantap'), ('gais', '')
    ]
    
    word_token = word_tokenize(data)
    word_list = []
    
    for word in word_token:
        for slang in slang_list:
            if word.lower() in slang[0].lower():
                word = slang[1]
                break
        word_list.append(word)
    
    data = ' '.join(word_list)
    return data

# Memproses semua fungsi menjadi satu
def proses(data):
    data = case_folding(data)
    data = stopword_cleaner(data)
    data = slang_cleaner(data)
    return data
```

Setelah semua fungsi dijalankan, saya cukup memanggil fungsi proses untuk membersihkan setiap tweet dan memasukkannya ke dataframe

```python
# Membaca file data mentah
df_mentah = pd.read_csv('data_mentah.csv')

# Membuat kolom baru clean tweet yang isinya sama dengan tweet
df_mentah['clean_tweet'] = df_mentah['tweet']

# Proses pembersihan tweet
for index in range(len(df_mentah)):
    df_mentah['clean_tweet'][index] = proses(df_mentah['clean_tweet'][index])

# Menyimpan ke file csv baru
df_mentah.to_csv('data_bersih.csv', index = False)

df_mentah
```

Berikut ini akan kita cek kata - kata apa saja yang paling banyak disebut oleh netizen. Fungsi yang dipakai berikut akan mereturn top words yaitu kata - kata terbanyak dan tweets yang berisi list yang mengekstraksi semua kata pada data.

```python
# Membuat fungsi yang plot distribusi kata, mereturn 20 kata paling banyak disebut dan list semua kata
def distribusi_kata(data):
    all_tweet = data.apply(lambda x: word_tokenize(str(x)))
    tweets = [word for tweet in all_tweet for word in tweet]
    fqdist = FreqDist(tweets)
    most_common_word = fqdist.most_common(20)
    fqdist.plot(20, cumulative=False, title = 'Distribusi kata')
    plt.show()
    return most_common_word, tweets

top_word, tweets = distribusi_kata(df_bersih['clean_tweet'])
print(top_word)
```

## Sentiment Analysis

Sebelum mengecek ekstraksi topik, saya akan melakukan analisa sentimen terlebih dahulu. Analisa ini digunakan untuk mengecek apakah tweet yang dibahas bersifat positif atau negatif. Pada kesempatan kali ini, saya menggunakan file kata positif dan negatif yang telah disediakan sebelumnya.

```python
# Membuka file kata positif dan negatif
pos_list= open("./kata_positif.txt","r")
pos_kata = pos_list.readlines()
neg_list= open("./kata_negatif.txt","r")
neg_kata = neg_list.readlines()  

# List kosong hasil untuk nanti di append dengan hasil dari analisa
hasil = []

# List kata - kata negasi yang mengubah sentimen
list_anti = ['tidak','lawan','anti', 'belum', 'belom', 'tdk', 'jangan', 'gak', 'enggak', 'bukan', 'sulit', 'tak', 'sblm']

# Menjadikan df clean tweet menjadi list
items = list(df_bersih['clean_tweet'])

# Loop untuk setiap tweet
for item in items:
    print(item.strip())
    tweets = item.strip().split() # Tokenize tweet
    
    count_p = 0 #nilai positif
    count_n = 0 #nilai negatif
    
    for tweet in tweets:
        for kata_pos in pos_kata: # mengecek tiap kata positif 
            if kata_pos.strip().lower() == tweet.lower(): # jika kata positif di temukan
                if items[items.index(item)-1] in list_anti: # cek terlebih dahulu apakah kata sebelumnya memuat negasi
                    print(items[items.index(item)-1], kata_pos, ['negatif']) # jika iya maka berubah jadi negatif
                    count_n += 1 # menambah perhitungan negatif
                else:
                    print(kata_pos, ['positif']) # jika tidak memuat negasi maka positif
                    count_p += 1 # menambah perhitungan positif
        for kata_neg in neg_kata: # cara pengecekan kata negatif serupa dengan loop positif di atas
            if kata_neg.strip().lower() == tweet.lower():
                if items[items.index(item)-1] in list_anti:
                    print(items[items.index(item)-1], kata_neg, ['positif'])
                    count_p += 1
                else:
                    print(kata_neg, ['negatif'])
                    count_n += 1
    
    hasil.append(count_p - count_n) # masukkan selisih positif negatif ke hasil
    print ("positif: "+str(count_p))
    print ("negatif: "+str(count_n))
    print ("----------------------------------")

sentiments = ["positif" if sentimen >= 0 else "negatif" for sentimen in hasil] # mengubah hasil menjadi sentimen
```

Kemudian, hasil analisa tersebut kita masukkan ke kolom baru.

```python
df_bersih['sentiment'] = sentiments
df_bersih
```

```python
# menghitung nilai mean dan std dari list hasil yang sudah dibuat sebelumnya
print ("Nilai rata-rata: "+str(np.mean(hasil)))
print ("Standar deviasi: "+str(np.std(hasil)))

# membuat plot setiap nilai tweet yang unik dan menghitung semua tweet yang bersesuaian
labels, counts = np.unique(hasil, return_counts=True)
plt.bar(labels, counts, align='center')
plt.gca().set_xticks(labels)
plt.show()
```

Jika kita lihat dari nilai rata - rata sentimen, tweet masyarakat cenderung ke arah positif. Sehingga kemungkinan besar masyarakat mendukung dengan adanya ppkm di natal dan tahun baru tersebut.

## Model TF-IDF

Pada sub bagian berikut ini, saya ingin mencoba melatih model machine learning menggunakan hasil sentimen sebelumnya. Langkah pertama yang saya lakukan adalah mereplace sentimen negatif dengan 0 dan positif dengan 1.

```python
# Membuat dictionary berisi isi yang akan diubah
ubah = {'negatif' : 0, 'positif' : 1}

# Mengcopy data ke dataframe baru kemudian mereplace data yang akan diganti
df_model = df_bersih.copy()
df_model.replace(ubah, inplace = True)
df_model
```

Setelah data sudah diubah, langkah selanjutnya adalah memanggil algoritma machine learning yang kita gunakan. Dalam kasus ini saya menggunakan TfidfVectorizer yang akan meng-vektorkan setiap kata yang ada pada tweet dan untuk library model yang dipakai adalah GaussianNB. Kemudian ubah hasil transform ke bentuk array karena nantinya akan dijadikan features atau x variabels. Sedangkan untuk label dari machine learning ini adalah sentiment.

```python
# Pemanggilan vectorizer
vectorizer = TfidfVectorizer (max_features=2000)

# Pemanggilan model
model_g = GaussianNB()

# Transformasi data
v_data = vectorizer.fit_transform(df_model['clean_tweet']).toarray()
print(vectorizer.fit_transform(df_model['clean_tweet']))
```

Angka pertama yang ada dalam kurung melambangkan urutan kalimat / tweet, angka kedua melambangkan urutan kata, dan angka di luar kurung adalah pembobotan

Setelah data divektorkan, kita bagi data menjadi data train dan data test, untuk menguji akurasi dari model yang kita buat nantinya. Kemudian model di belajarkan pada data train dan memprediksi data test. Setelah itu kita cek tingkat akurasinya dengan label pada data test.

```python
# Pembagian data menjadi train dan test
X_train, X_test, y_train, y_test = train_test_split(v_data, df_model['sentiment'], test_size=0.2, random_state=18)

# Pembelajaran model ke data train
model_g.fit(X_train,y_train)

# Prediksi model terhadap data uji
y_preds = model_g.predict(X_test)

# Confusion Matrix
print('Confusion Matrix :')
print(confusion_matrix(y_test,y_preds))
```

Pada hasil confusion matrik kita dapat :
1. Nilai True Positif yaitu apabila model memprediksi positif dan itu benar adanya. Model berhasil memprediksi 6 dari 10 klasifikasi 0.
2. Nilai False Positif yaitu apabila model memprediksi positif padahal aslinya negatif. Model gagal memprediksi 5 dari 30 untuk klasifikasi 1.
3. Nilai True negatif yaitu apabila model memprediksi negatif dan benar. Model berhasil memprediksi 25 dari 30 klasifikasi 1.
4. Nilai False negatif yaitu apabila model memprediksi negatif padahal aslinya positif. Model gagal memprediksi 4 dari 10 klasifikasi 0.

Nilai tersebut kemudian dihitungkan kembali dan menjadi classification report yang terdiri atas
1. Precision = perbandingan nilai benar dengan semua prediksi pada label tersebut seperti contoh untuk klasifikasi 0 didapat TP/(TP+FP) = 6/(6+5) = 0.55
2. Recall = perbandingan nilai benar dengan semua prediksi yang aslinya mengarah pada label tersebut yaitu TP/(TP+FN) = 6/(6+4) = 0.6
3. F1-score = harmoni antara precision dan recall yaitu 2.P.R/(P+R) = 2.0.55.0.6/(0.55+0.6) = 0.57
4. Support = jumlah observasi tiap label
5. Accuracy = perbandingan nilai benar dengan seluruh prediksi yaitu (TP+TN)/(TP+FP+TN+FN) = 6+25/6+5+25+4 = 31/40 = 0.775

Dari nilai accuracy 0.775 bisa dikatakan akurasi model cukup baik dan bisa digunakan. Untuk lebih jelasnya berikut hasil lengkap perhitungan klasifikasi tersebut.

## Grafik Networkx

Pada sub bagian ini saya ingin melihat hubungan antar kata menggunakan grafik networkx. Langkah pertama adalah dengan membuat data frame bigram yang kita filter untuk count yang lebih dari 1

```python
# Ambil setiap kata pada tweet dan dijadikan list
tweet_gram = [word for tweet in df_model["clean_tweet"] for word in tweet.split()]

# Ambil pasangan dua kata / distribusi bigram
fdist = nltk.FreqDist(nltk.ngrams(tweet_gram, 2))

# Items dari distribusi bigram memiliki 2 value yaitu tuple berisi dua komponan yang akan menjadi nodes dan count yang menjadi
# weight
list_data = [[k,v]for k,v in fdist.items()]

# Mengubah list dari bigram menjadi dataframe
df_bigram = pd.DataFrame(list_data, columns=['bigram', 'count'])

# Memfilter bigram yang lebih dari 1
lebih = df_bigram['count'] > 1
df_bigram_sort = df_bigram[lebih]

# Mengurutkan hasil dataframe dan tampilkan
df_bigram_sort = df_bigram_sort.sort_values(by=["count"], ascending=False).copy().reset_index(drop=True)

df_bigram_sort
```

Setelah dif bigram dibuat, data kemudian di transpose terlebih dahulu. Kemudian lakukan inisiasi object networkx dan ambil nodes dan weight nya. Kemudian buat list degree dari node kita.

```python
# Transpose data
d = df_bigram_sort.set_index("bigram").T.to_dict("records")

# Inisiasi object networkx
G = nx.Graph()

# Memasukkan relasi node dan weight
for k,v in d[0].items():
    G.add_edge(k[0], k[1], weight=(v*10))
    
# Degree = berapa banyak relasi yang terhubung dengan node kita
deg = list(G.degree)

# Menunjukkan degree yang sudah diurutkan
m = [data[0] for data in deg]
n = [data[1] for data in deg]
data_degree = pd.DataFrame({'kata' : m, 'degree' : n})
data_degree_sort = data_degree.sort_values(by=["degree"], ascending=False).copy().reset_index(drop=True)
data_degree_sort
```

Selanjutnya kita plot semua node ke dalam grafik sebagai berikut.

```python
# Plot networks
fig, ax = plt.subplots(figsize=(50, 50))
pos = nx.spring_layout(G, k=2.5)
nx.draw_networkx(G, pos,
        font_size=12,
        width=1,
        edge_color='grey',
        node_color='purple',
        with_labels = False,
        node_size = [v[1] * 100 for v in deg],
        ax=ax)

# Create offset labels
for key, value in pos.items():
    x, y = value[0]+.025, value[1]+.015
    ax.text(x, y,
            s=key,
            bbox=dict(facecolor='grey', alpha=0.25),
            horizontalalignment='center', fontsize=12)
    
plt.show()
# G - graph kita
# Pos - dictionary koordinat dari setiap node
# node_size - mengubah besaran node sesuai kebutuhan
```

Dapat kita lihat dari df degree sebelumnya bahwasanya kata - kata terbanyak yang terhubung ke kata lain adalah ppkm dengan 34 degree, level: 10 degree, melaksanakan: 8 degree, dan himbauan: 7 degree. Sehingga kebanyakan pembahasan mengenai himbauan pelaksanaan ppkm

## Topic Extraction

Sekarang kita akan melakukan topic extraction menggunakan lda model. Sebelumnya kita akan buat data ppkm yang berisi dari tokenize semua tweet dan membuat mapping dalam satu dokumen

```python
# Tokenize semua tweet
data_ppkm = [data.split() for data in df_bersih["clean_tweet"]]

# Mapping kata dalam satu dokumen
dictionary = corpora.Dictionary(data_ppkm)
doc_term_matrix = [dictionary.doc2bow(doc) for doc in data_ppkm]
```

Pada model LDA ada beberapa parameter yang harus diisi diantaranya adalah topic, alfa dan betha. 
Alpha - parameter similarity data. 
- Nilai Alpha yang rendah merepresentasikan dokumen yang memiliki topik yang mudah di bedakan. Nilai lebih tinggi merepresentasikan dokumen/data yang bisa memiliki lebih banyak kemungkinan
- Beta - parameter similarity topik. Ditunjukan dengan distribusi kata per topik. Nilai beta yang rendah berarti semakin banyak kata unik yang merepresentasikan suatu topik. Nilai yang lebih tinggi, artinya topik memiliki lebih banyak common word.

Pada percobaan pertama mari kita coba isi untuk nilai topic nya adalah 3. 

```python
Lda = gensim.models.ldamodel.LdaModel
ldamodel = Lda(
    random_state=100,
    chunksize=100,
    per_word_topics=True,
    corpus = doc_term_matrix, 
    num_topics=3, 
    id2word = dictionary, 
    passes=50
    )

pprint(ldamodel.print_topics())
```

Karena num topik yang kita pilih adalah 3 maka indeks awal pada tuple tersebut melambangkan indeks topik kita (0-2). Angka di sebelah kiri kata adalah bobot dari topik yang bersangkutan. Pada model ini kata yang ditampilkan adalah 10 kata dengan bobot terbesar.

Jumlah topik 3 ini belum tentu yang paling optimal. Maka dari itu kita cek coherence valuenya terlebih dahulu dengan cara berikut. Coherence value menunjukkan nilai jarak dari setiap kata pada suatu topik. Nilainya berada di antara 0 - 1

```python
coherence_model_lda = CoherenceModel(model=ldamodel, texts=data_ppkm, dictionary=dictionary, coherence='c_v')
coherence_lda = coherence_model_lda.get_coherence()
print('\nCoherence Score: ', coherence_lda)
```

Kita peroleh coherence value sebesar 0.434163708537741 namun kita belum mengetahui apakah ada parameter num topic yang lebih baik. Dengan melakukan hyper tuning parameter kita bisa mencari coherence value terbaik.

```python
# Fungsi hyper tuning pada model gensim
def compute_coherence_values(corpus, dictionary, k, a, b):
    lda_model = gensim.models.LdaMulticore(corpus=doc_term_matrix,
                                           id2word=dictionary,
                                           num_topics=k, 
                                           random_state=100,
                                           chunksize=100,
                                           passes=10,
                                           alpha=a,
                                           eta=b)
    coherence_model_lda = CoherenceModel(model=lda_model, texts=data_ppkm, dictionary=dictionary, coherence='c_v')
    return coherence_model_lda.get_coherence()
```

Setelah itu kita lakukan proses pencarian nilai koheren terhadap parameter yang dimiliki. karena itu kita kan mencari nilai k, a, dan b terbaik.

```python
# Topics range
min_topics = 2
max_topics = 11
step_size = 1
topics_range = range(min_topics, max_topics, step_size)
# Alpha parameter
alpha = list(np.arange(0.01, 1, 0.3))
alpha.append('symmetric')
alpha.append('asymmetric')
# Beta parameter
beta = list(np.arange(0.01, 1, 0.3))
beta.append('symmetric')
# Validation sets
num_of_docs = len(doc_term_matrix)
corpus_sets = [gensim.utils.ClippedCorpus(doc_term_matrix, num_of_docs*0.75), 
               doc_term_matrix]
corpus_title = ['75% Corpus', '100% Corpus']
model_results = {'Validation_Set': [],
                 'Topics': [],
                 'Alpha': [],
                 'Beta': [],
                 'Coherence': []
                }
# Can take a long time to run
if 1 == 1:
    pbar = tqdm.tqdm(total=540)
    
    # iterate through validation corpuses
    for i in range(len(corpus_sets)):
        # iterate through number of topics
        for k in topics_range:
            # iterate through alpha values
            for a in alpha:
                # iterare through beta values
                for b in beta:
                    # get the coherence score for the given parameters
                    cv = compute_coherence_values(corpus=corpus_sets[i], dictionary=dictionary, 
                                                  k=k, a=a, b=b)
                    # Save the model results
                    model_results['Validation_Set'].append(corpus_title[i])
                    model_results['Topics'].append(k)
                    model_results['Alpha'].append(a)
                    model_results['Beta'].append(b)
                    model_results['Coherence'].append(cv)
                    
                    pbar.update(1)
    pd.DataFrame(model_results).to_csv('lda_tuning_ppkm_results.csv', index=False)
    pbar.close()
```

Kemudian hasil dari tuning kita jadikan data frame dan csv untuk dianalisa lebih lanjut. Validasi set yang dipilih adalah 75% dengan 100%
